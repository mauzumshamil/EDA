```python
# so we can find the skewness in some and some not
#skewed= al ,su ,bgr ,bu ,sc , pot  soo its median
#not skewed= age ,bp ,sg ,sod ,hemo soo its mean
```

```python
for i in ['al','su','bgr','bu','sc','pot']:
  df[i]=df[i].fillna(df[i].median())
df[i]
```

```python
for i in ['age' ,'bp' ,'sg' ,'sod' ,'hemo']:
  df[i]=df[i].fillna(df[i].mean())
df[i]
```

```python
Q1=np.percentile(df['pot'],25,interpolation='midpoint')
Q2=np.percentile(df['pot'],50,interpolation='midpoint')
Q3=np.percentile(df['pot'],75,interpolation='midpoint')

print(Q1)
print(Q2)
print(Q3)

IQR=Q3-Q1
IQR

upper_limit=Q3+1.5*IQR
lower_limit=Q1-1.5*IQR

outlier=[]
for x in df['pot']:
  if((x>upper_limit) or (x<lower_limit)):
    outlier.append(x)

outlier
```

```python
df1=df['pot']>upper_limit
df2=df['pot']<lower_limit
print(df1)
print(df2)
```
