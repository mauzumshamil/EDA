```python
#graphs are skewed so use median to fill na values
Skewness is a measurement of the distortion of symmetrical distribution or asymmetry in a data set.
Skewness is demonstrated on a bell curve when data points are not distributed symmetrically to the left and right
sides of the median on a bell curve.
```

```python
# fillna()-used to fill null value in the dataframe
# df[i]=df[i].fillna(df[i].median())
              #  or
# df['beer_servings'].fillna(df['beer_servings'].median())

for i in ['beer_servings', 'spirit_servings', 'wine_servings',
       'total_litres_of_pure_alcohol']:
       df[i]=df[i].fillna(df[i].median())

df[i]
```
